```python
from procustes_utils import create_test_set,draw_shapes
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
```

```python
canvas, triangles = create_test_set()
```

```python
def draw_triangles(triangles):
    new_triangles = []

    for triangle in triangles:
        trian = [[triangle[i], triangle[i+1]] for i in range(0, 5, 2)]
        new_triangles.append(np.array(trian))
    
    plt.figure()
    for triangle in new_triangles:

        plt.scatter(triangle[:, 0], triangle[:, 1])

        t1 = plt.Polygon(triangle, color=np.random.rand(3,))
        plt.gca().add_patch(t1)


    plt.show()
```

```python
draw_triangles(triangles)
```

```python
from procustes_utils import get_translation,procrustes_analysis
```

```python

#get translation of reference landmark
x, y = get_translation(triangles[0])
 #create array for new shapes, append reference shape to it
new_shapes = []
new_shapes.append(triangles[0])

    #superimpose all shapes to reference shape
for i in range(1, 5):
    new_shape = procrustes_analysis(triangles[0], triangles[i])
    new_shape[::2] = new_shape[::2] + x
    new_shape[1::2] = new_shape[1::2] + y
    new_shapes.append(new_shape)
```

```python

draw_triangles(new_shapes)
```
