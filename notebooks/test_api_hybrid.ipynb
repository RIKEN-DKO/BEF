```python
import requests
import json
```

```python
base_url = 'http://localhost:5555'  # Replace with your server's address

# Querying the 'search' endpoint
search_data = {
    'event_type': 'id',
    'query': 'disease',
    'num_res': 20,
    'alpha': 0.5
}

response = requests.post(f'{base_url}/search', json=search_data)
search_results = response.json()
print(json.dumps(search_results, indent=4))  
```

```python
# Querying the 'get_doc_annotations' endpoint
event_type = 'cg'
doc_id = '4460956'
response = requests.get(f'{base_url}/annotations/{event_type}/{doc_id}')
annotations = response.json()
print(json.dumps(annotations, indent=4)) 
```
